```python
import pandas as pd
import numpy as np

#read the dataset from dataset folder "Individual Traffic Data.csv" & "Traffic Statistics Data.csv"
_df_individual = pd.read_csv("Dataset/Individual Traffic Data.csv", sep=';', engine='python');
_df_overall = pd.read_csv("Dataset/Traffic Statistics Data.csv", sep=';', engine='python');

#get the header of both dataframe
print("This is the list of _df_individual: " + str(list(_df_individual)))
print("This is the list of _df_overall: " + str(list(_df_overall)))
```

```python
_df_individual
```

```python
#average sum of 'Avg. Speed' (whole column of _df_individual)
_avg_sum_spd_individual = _df_individual[' Speed'].abs().mean()
print(_avg_sum_spd_individual)

#average sum of 'Avg. Speed' (based on selected element in _df_individual)
#_avg_sum_spd_individual = _df_individual.groupby([' Speed']).size().reset_index(name='counts')
_avg_sum_spd_individual_nb = _df_individual[(_df_individual[' Speed'].values <=0) & (_df_individual['UMRRID'] == 0)]
print(_avg_sum_spd_individual_nb[' Speed'].abs().mean())

_avg_sum_spd_individual_sb = _df_individual[(_df_individual[' Speed'].values >0) & (_df_individual['UMRRID'] == 0)]
print(_avg_sum_spd_individual_sb[' Speed'].abs().mean())
```
